```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error
```

```python
# Load the dataset
merged_df1.info()
```

```python
# Drop rows where 'OrderDate' is missing
merged_df1 = merged_df1.dropna(subset=['OrderDate'])
```

```python
merged_df1
```

```python
# Group by year and month to get monthly sales data
monthly_sales = merged_df1.groupby(['year', 'month'])['Total_sales'].sum().reset_index()
```

```python
monthly_sales 
```

```python
# Create a 'Date' column for the time series
monthly_sales['Date'] = pd.to_datetime(monthly_sales[['year', 'month']].assign(DAY=1))
```

```python
monthly_sales
```

```python
# Set 'Date' as the index
monthly_sales.set_index('Date', inplace=True)
```

```python
monthly_sales
```

```python
# Drop unnecessary columns
monthly_sales = monthly_sales.drop(['year', 'month'], axis=1)
```

```python
monthly_sales
```

```python
 #Plot the sales data
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, label='Monthly Sales')
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
```

```python
# Fit the ARIMA model
model = ARIMA(monthly_sales['Total_sales'], order=(1, 1, 1))  
arima_result = model.fit()

# Forecast for the next 12 months
forecast = arima_result.forecast(steps=12)

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, label='Actual Sales')
plt.plot(forecast, label='Forecasted Sales', linestyle='--')
plt.title('Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
